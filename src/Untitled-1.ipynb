```python
from pathlib import Path
import json
import pymupdf
import base64
from pdf_seperator import PDFSeperator

data = Path("./output.json")
data = json.loads(data.read_text())
b64_pdf = data["parsed"][0]["pdf_bytes"]
raw_pdf_bytes = base64.b64decode(b64_pdf)

page_range = data["parsed"][0]["data"]["page_range"]

pages_bytes = PDFSeperator(pdf_bytes=raw_pdf_bytes).extract_page_range(page_range["start_page"],page_range["end_page"])

with open("output.pdf", "wb") as f:
    f.write(raw_pdf_bytes)
with open("output_parsed.pdf", "wb") as f:
    f.write(pages_bytes)
```
